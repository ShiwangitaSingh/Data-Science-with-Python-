```python
# Filling missing values with a default value
df_filled = df.fillna(0)
```

```python
# Filling missing values with the last valid value before it in the same column.
df_filled2 = df.ffill()
```

```python
# Filling missing values with the next valid value after it in the same column.
df_filled3 = df.bfill()
```

- **Data Modification**

```python
# Shows all the columns 
df_filled.columns
```

```python
# Deleting a Column
df_filled.drop(['Observation'], axis=1, inplace=True)
```

```python
# Cheking columns again
df_filled.columns
```

- **Data Analysis**

```python
# Calculating the Interquartile Range (IQR)
Q1 = df_filled.quantile(0.25)
Q3 = df_filled.quantile(0.75)
IQR = Q3 - Q1
IQR
```

```python
# Calculating Outliers and removing them
upper_bound = Q3 + 1.5*IQR
lower_bound = Q1 - 1.5*IQR
# outliers = df_filled[((df_filled < lower_bound)|(df_filled > upper_bound)).any(axis=1)]
# df_filled = df_filled.drop(outliers.index) ........................OR....................
df_filled = df_filled[~((df_filled < lower_bound)|(df_filled > upper_bound)).any(axis=1)]
df_filled
```
